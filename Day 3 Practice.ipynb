```python
stocks = ['ICIC', 'HDFC','Reliance', 'ITC', 'SBIN']
price = [361.45,1073.45,1784.9,207.55,184.70]

len1 = len(stocks)

stocks_range = np.arange(0,len1)

plt.bar(stocks_range, price)

plt.xticks(stocks_range,stocks)
plt.xlabel("Nifty 50 Stocks ")
plt.ylabel('Last traded price')
plt.title('Stock performance')
plt.show()

```

```python
plt.barh(stocks_range, price)
plt.yticks(stocks_range,stocks)
plt.ylabel("Nifty 50 Stocks ")
plt.xlabel('Last traded price')
plt.title('Stock performance')
plt.show()

```

```python
a = np.array([2,5,7,8,1])

s = pd.Series(a,index=['a','b','c','d','e'])

print(s)

print("The number of dimensions of Series are : ",s.ndim)

print("The number of bytes used by Series are : ",s.nbytes)

print("The shape of Series is : ",s.shape)

print("The number of element in Series are : ",s.size)

print("The values of Series are : ",s.values)

print("The index of Series are : ",s.index)
```

```python
a = np.array([3,5,8,np.NaN,9])

s1 = pd.Series(a)

s1 = s1 + 5

print("Addition of two Series is :\n",s1)
```

```python
a = np.array([3,5,8,np.NaN,9])
b = np.array([2,6,8,11,5])

s1 = pd.Series(a)
s2 = pd.Series(b)


s1 = s1.add(s2,fill_value=0)

print("The addition is :\n", s1)
```

```python
a = np.random.randint(1,10,(4,2))
print("Original array\n",a,"\n")

print("added by 2 \n",a + 2)

print("sub by 6\n",a - 6)

print("multipy by 2.5\n",a * 2.5)

print("divided\n",a / 4)

print("reminder\n",a % 2)

print("floor division\n",a // 4)
```

```python
a=np.arange(1,10).reshape((3,3))
print("first array is : ",a)
b=np.arange(11,20).reshape((3,3))
print("\n second array is :",b)
c=np.concatenate((a,b))
print("\n concatenated array along axis=0 is :\n",c)
c=np.concatenate((a,b),axis=1)
print("\n concatenated array along axis=1 is :\n",c)
```

```python
a=np.random.randint(1,10,(3,3))
b=np.random.randint(10,20,(3,3))

print(" first array is :",a)
print("\n second array is :",b)
c=np.hstack((a,b))
print("horizontally stacked array :",c)
```

```python
a=np.random.randint(1,10,(3,3))
b=np.random.randint(10,20,(3,3))
print(" first array is :",a)
print("\n  second array is :",b)
c=np.vstack((a,b))
print(" horizontally stacked array  :",c)
```

```python
a=np.arange(20)
a=np.reshape(a,(2,10))
print(" original array is :",a)
x,y,z = np.split(a,[2,5],axis=1)
print("\n first array ",x)
print("\n second array ",y)
print("\n  third array ",z)
```
